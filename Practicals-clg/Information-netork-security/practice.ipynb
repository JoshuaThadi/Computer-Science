## Caesar Cipher

```python
# code to encrypt the plaintext into caesar cipher text
def encrypt(text,s):
    result = ""
    for i in range(len(text)):
        char = text[i]
        
        if char.isupper():
            result = result + chr((ord(char) + s - 65) % 26 + 65)
        else:
            result = result + chr((ord(char) + s - 97) % 26 + 97)
    return result
text = input("Enter the text to encrypt: ")
s = int(input("Enter the shift: "))
print("Text: " + text)
print("Shift: "+ str(s))
print("CipherText: "+ encrypt(text, s))
```

```python
# Code to decrypt the caesar cipher text into plaintext
# Python code to decrypt Caesar Cipher text

def decrypt(cipher_text, shift):
    result = ""
    
    for char in cipher_text:
        # Check if character is uppercase
        if char.isupper():
            result += chr((ord(char) - shift - 65) % 26 + 65)
        # Check if character is lowercase
        elif char.islower():
            result += chr((ord(char) - shift - 97) % 26 + 97)
        else:
            # Keep non-alphabetic characters as they are (spaces, punctuation, etc.)
            result += char
    
    return result

# Input section
cipher_text = input("Enter the cipher text: ")
shift = int(input("Enter the shift number: "))

# Decryption process
print("Decrypted Text:", decrypt(cipher_text, shift))
```

## Railfence Cipher

```python
# Code to implement the railfence cipher
# encrypting the message
def railfence(string):
    result = ""
    for i in range(len(string)):
        if i % 2 == 0:
            result = result + string[i]
    for i in range(len(string)):
        if i % 2 != 0:
            result = result + string[i]
    return result
string = input("Enter the text to encrypt: ")
print("Cipher: " + railfence(string))
```

```python
# decrypting the message
# Python code to decrypt Rail Fence Cipher (for 2 rails)

def decryptRailFence(cipher):
    # Find the length of each rail
    mid = len(cipher) // 2
    if len(cipher) % 2 != 0:
        mid += 1  # for odd length strings
    
    # First half -> even indices
    # Second half -> odd indices
    first = cipher[:mid]
    second = cipher[mid:]
    
    # Reconstruct the original text
    result = ""
    for i in range(len(second)):
        result += first[i] + second[i]
    if len(first) > len(second):
        result += first[-1]
        
    return result


# Input section
cipher = input("Enter the cipher text to decrypt: ")
print("Decrypted text:", decryptRailFence(cipher))
```

## SHA1 algorithm - MAC

#### SHA (Secure Hash Algorithm)

**SHA (Secure Hash Algorithm)** is a family of cryptographic hash functions designed by the **U.S. National Security Agency (NSA)** and published by **NIST (National Institute of Standards and Technology)**.

These algorithms are used to convert data (of any size) into a **fixed-size hash value** — a unique digital fingerprint.

A hash function takes an input (for example: a message, password, or file) and produces a fixed-size hash value that:

1. **Represents the data uniquely**  
2. **Is irreversible** — you cannot recover the original input from the hash  
3. **Exhibits the avalanche effect** — the hash changes completely if the input changes even slightly

SHA-1 (Secure Hash Algorithm 1) is a cryptographic hash function designed by the U.S. National Security Agency (NSA) and published by NIST in 1995.

It converts any input data (like text or files) into a 160-bit (20-byte) hash value — usually represented as a 40-character hexadecimal string.

```python
# implementing SHA! algorithm
# code to encode
import hashlib
str = input("The value of encode: ")
result = hashlib.sha1(str.encode())
print("The hexadecimal equivalent of SHA1 is: ")
print(result.hexdigest())
```

## MD5 algorithm - MAC

MD5 (Message Digest Algorithm 5) is a cryptographic hash function developed by Ronald Rivest in 1991.

MD5 takes an input (any length of data) and produces a 128-bit (16-byte) fixed-length output, usually represented as a 32-character hexadecimal number.

```python
# code to implement MD5
import hashlib
result = hashlib.md5(b"Ismile")
result1 = hashlib.md5(b"Esmile")

# code to encrypt
print("The byte equivalent of hash is : ", end="")
print(result.digest())
print("The byte equivalent of hash is : ", end="")
print(result1.digest())
```

## Diffe-Hellman Algorithm

The **Diffie-Hellman (DH) algorithm** is a **cryptographic key exchange protocol** that allows two parties to **securely share a secret key** over an insecure channel **without transmitting the key itself**.

It is **not used for encrypting messages directly** — it is used to generate a **shared secret key** for symmetric encryption.

---

#### Public Parameters
Both parties agree on:
- A **large prime number** `p`  
- A **generator** `g` (primitive root modulo `p`)  

> These values are **public**.

---

#### Private Keys
Each party chooses a **secret private key**:
- **Alice** chooses `a`  
- **Bob** chooses `b`


```python
# Diffe-Hellman algorithm
from random import randint
if __name__ == "__main__":
    G = 9
    P = 23
    
    print("The value of G is :%d"%(G))
    print("The value of P is :%d"%(P))
    
    a = 4
    print('Secret number for alice is :%d'%(a))
    x = int(pow(G,a,P))
    
    b = 6
    print('Secret number for Bob is :%d'%(b))
    y = int(pow(G,b,P))
    
    ka = int(pow(x,b,P))
    kb = int(pow(y,a,P))
    
    print('Secret key for the Alice is : %d'%(ka)) 
    print('Secret Key for the Bob is : %d'%(kb))
```
